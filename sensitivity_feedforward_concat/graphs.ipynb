# Sensitivity analysis for the concatenated feedforward model

# Incoherent feedforward loop concatenated

$$\frac{\mathrm{d} I_{1}}{\mathrm{d} t} = S*k_{Ia1}(1-I_{1})-k_{Ii1}I_{1}$$

$$\frac{\mathrm{d} M_{1}}{\mathrm{d} t} = I_{1}*k_{Ma1}(1-M_{1})-k_{Mi1}M_{1}$$

$$\frac{\mathrm{d} R_{1}}{\mathrm{d} t} = I_{1}*k_{Ra1}(1-R_{1})-M_{1}*k_{Ri1}\frac{R_{1}}{R_{1} + K_{1}}$$

$$\frac{\mathrm{d} I_{2}}{\mathrm{d} t} = R_{1}*k_{Ia2}(1-I_{2})-k_{Ii2}I_{2}$$

$$\frac{\mathrm{d} M_{2}}{\mathrm{d} t} = I_{2}*k_{Ma2}(1-M_{2})-k_{Mi2}M_{2}$$

$$\frac{\mathrm{d} R_{2}}{\mathrm{d} t} = I_{2}*k_{Ra2}(1-R_{2})-M_{2}*k_{Ri2}\frac{R_{2}}{R_{2} + K_{2}}$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ldata(archive):
    f=open(archive)
    data=[]
    for line in f:
        e=line.strip()
        col=line.split()
        data.append(col)	
    return data
```

```python
data=ldata("sensitivity_feedforward.txt")
data_float=[]
for i in data:
  data_float.append([float(i[0]), float(i[1])])
```

```python
positive_data = []
negative_data = []
for i in data_float:
    positive_data.append(i[0])
    negative_data.append(i[1])
```

```python
# some example data
threshold = 0.0

x= [r"$k_{Ia1}$", r"$k_{Ii1}$",r"$k_{Ma1}$",r"$k_{Mi1}$",r"$k_{Ra1}$",r"$k_{Ri1}$",r"$K_{R1}$",r"$k_{Ma2}$",r"$k_{Mi2}$",r"$k_{Ra2}$",r"$k_{Ri2}$",r"$K_{R2}$",r"$k_{Ia2}$",r"$k_{Ii2}$"]


# and plot it
fig, ax = plt.subplots()

ax.bar(x, negative_data, width=0.5, color='grey')
ax.bar(x, positive_data, width=0.5, color='grey')

# horizontal line indicating the threshold
ax.plot([-0.5, 14.5], [threshold, threshold], 'w', lw="1.5")
plt.hlines(0.25, -0.5, 14.5, color='grey', alpha=0.5)
plt.hlines(0.5, -0.5, 14.5, color='grey', alpha=0.5)
plt.hlines(0.75, -0.5, 14.5, color='grey', alpha=0.5)
plt.hlines(1.0, -0.5, 14.5, color='grey', alpha=0.5)
plt.hlines(-0.25, -0.5, 14.5, color='grey', alpha=0.5)
plt.hlines(-0.5, -0.5, 14.5, color='grey', alpha=0.5)
plt.hlines(-0.75, -0.5, 14.5, color='grey', alpha=0.5)
plt.hlines(-1.0, -0.5, 14.5, color='grey', alpha=0.5)
plt.xticks(rotation = 45)
plt.ylabel(r"$log_{10}$ perturbation")
plt.xlim((-0.670,13.5))
plt.show()
```
